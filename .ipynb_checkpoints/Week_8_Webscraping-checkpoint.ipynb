#using the url below, perform the following
1. webscrape the financial website and load it in a Pandas dataframe
2. draw out three insights from the data

```python
url = "https://www.worlddata.info/africa/nigeria/inflation-rates.php"
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as npy
```

```python
r = requests.get(url)
html = r.content
soup = BeautifulSoup(html)
```

```python
soup
```

```python
html
```

```python
r
```

```python
table_header = []
for header in soup.find_all("th"):
    table_header.append(header.text)
```

```python
df = pd.DataFrame(columns = table_header)
df
```

```python
list_of_rows = []
table_row = soup.find_all("tr")
for i in range (1, len(table_row)):
    list_of_each_rows = []
    second_table = table_row[i]
    for row in second_table:
      list_of_each_rows.append(row.text)
    list_of_rows.append(list_of_each_rows)
    
for j in list_of_rows:
    #print(j)
    df.loc[len(df)] = j
    
    
```

```python
df
```

### Checking Column Names and Renaming them Properly

```python
df.columns
```

```python
new_column_names = ["Year", "NGN_Inflation_rate_(%)", "EU_Inflation_rate_(%)", "USA_Inflation_rate_(%)", "World_Inflation_rate_(%)"]
df.columns = new_column_names
df
```

```python
df.count()
```

```python
df.dtypes
```

### Checking for missing values

```python
df.isna().sum()
```

### Removing the % sign frown the columns

```python
for col in df.columns:
    if "%" in col:
        df[col] = df[col].apply(lambda x: x.replace("%", ""))
df
```

### Replace the "n/a" with "0" in the world_inflation_rate_(%) column

```python
df["World_Inflation_rate_(%)"] = df["World_Inflation_rate_(%)"].replace("n/a", "0")
df
```

### Convert columns to float datatypes

```python
for i in df.columns[1:]:
    df[i] = df[i].astype("float")
```
